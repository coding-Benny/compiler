## 11. Text Cleaning
- inaugural speeches in NLTK corpus

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt

from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.probability import FreqDist
```

```python
speechObama = nltk.corpus.inaugural.words('2009-Obama.txt')
```

### remove stop words

```python
stop_words = stopwords.words('english')
```

```python
puncs = [' ', ',', '.', ';', "'", '--', '-', ':']
for x in puncs:
    stop_words.append(x)
```

```python
n_Obama = []
for w in speechObama:
    if w not in stop_words:
        n_Obama.append(w)
```

```python
len(speechObama), len(n_Obama)
```

### visualize using matplotlib

```python
fdist = FreqDist(n_Obama)
fdist.most_common(20)
```

```python
fdist.plot(20, cumulative=False)
plt.show()
```

### 2013

```python
speechObama = nltk.corpus.inaugural.words('2013-Obama.txt')
n_Obama = []
for w in speechObama:
    if w not in stop_words:
        n_Obama.append(w)
fdist = FreqDist(n_Obama)
fdist.plot(20, cumulative=False)
plt.show()
```

### 2017

```python
speechTrump = nltk.corpus.inaugural.words('2017-Trump.txt')
n_Trump = []
for w in speechTrump:
    if w not in stop_words:
        n_Trump.append(w)
fdist = FreqDist(n_Trump)
fdist.most_common(10)
```

```python
fdist.plot(20, cumulative=False)
plt.show()
```

### More cleaning with lemmatization -> Normalization

```python
lemmatizer = WordNetLemmatizer()
lemma_speech = []
for x in n_Trump:
    lemma_speech.append(lemmatizer.lemmatize(x, 'v'))
    
fdist = FreqDist(lemma_speech)
fdist.most_common(10)
```
