# 고등교육법 일부개정 법률안

## 1. 필요한 라이브러리 import

```python
import nltk
import matplotlib.pyplot as plt
from konlpy.corpus import kobill
from konlpy.tag import Okt
from wordcloud import WordCloud, STOPWORDS
import stylecloud
from IPython.display import Image
```

- 한글 지원 코드

```python
import platform
from matplotlib import font_manager, rc
if platform.system()=='Windows':
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
```

## 2. 고등교육법 일부개정 법률안 내용 읽어보기

```python
ko_files = kobill.fileids()
doc = kobill.open('1809894.txt').read()
doc
```

## 3. 명사구 추출

```python
t = Okt()
tokens = t.nouns(doc)
tokens
```

## 4. 빈도수 처리

```python
ko = nltk.Text(tokens, name='대한민국 국회 의안 제 1809894호')

print(len(ko.tokens))
print(len(set(ko.tokens)))

ko.vocab() # frequency distribution
```

## 5. 빈도수 그래프

```python
plt.figure(figsize=(12, 6))
ko.plot(50)
plt.show()
```

## 6. 제외할 문자 처리

```python
stop_words = ['.', '(', ')', ',', "'", '%', '-', 'X', ').', '×','의','자','에','안','번',
              '호','을','이','다','만','로','가','를', '및', '발', '인', '면', '증', '수',
              '학', '위', '임', '정', '함', '제', '관', '부', '칙', '날', '령']

ko = [word for word in ko if word not in stop_words]
ko
```

## 7. 다시 빈도수 그래프 확인

```python
ko = nltk.Text(ko, name='대한민국 국회 의안 제 1809894호')

plt.figure(figsize=(12, 6))
ko.plot(50)
plt.show()
```

## 8. 특정 단어 빈도수 확인

```python
ko.count('학위')
```

## 9. 산점도 

```python
plt.figure(figsize=(12, 6))
ko.dispersion_plot(['학위', '수업', '이수'])
```

## 10. 주어진 단어의 주변부 확인

```python
ko.concordance('학사')
```

## 11. 연어 확인

```python
ko.collocations()
```

## 12. 빈도 수가 높은 100개의 단어로 word cloud 그리기

```python
data = ko.vocab().most_common(100)

# Windows : font_path='c:/Windows/Fonts/malgun.ttf'
# Mac : font_path='/Library/Fonts/AppleGothic.ttf'
wordcloud = WordCloud(font_path='./NanumSquareB.ttf',
                     relative_scaling = 0.2,
                     background_color = 'white',
                     ).generate_from_frequencies(dict(data))
plt.figure(figsize=(12, 8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

### stylecloud를 사용하여 예쁘게 출력하기

```python
stylecloud.gen_stylecloud(text=dict(data),
                          icon_name='fas fa-graduation-cap',
                          palette='colorbrewer.diverging.Spectral_11',
                          background_color='black',
                          gradient='horizontal', font_path='./NanumSquareB.ttf')
```
